```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
### New HDL UK/Japan Biobank trans-ethnic fine mapping set sizes
hdl_loci_df = pd.read_csv('locus_set_sizes.txt', sep='\t', index_col=0)
hdl_loci_df
```

```python
sns.set_context("talk", font_scale=0.8)
figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')
plt.ylabel='Causal Set Size'
ax = sns.boxplot(data=hdl_loci_df, palette='muted')
#ax.yaxis.grid(True) # Show the horizontal gridlines
# sns.swarmplot(data=hdl_loci_df, palette='muted')
ax.set(xlabel="", ylabel="Causal Set Size")
# plt.xticks(rotation=0)
plt.savefig('hdl_box_and_swarm.png', bbox_inches='tight')
```

```python
# Scatter plots HDL
sns.set_context("talk", font_scale=1.0)
figure(num=None, figsize=(18, 4), dpi=80, facecolor='w', edgecolor='k')
plt.subplots_adjust(wspace = 0.3)

plt.subplot(1, 3, 1)
ax1 = sns.scatterplot(data=hdl_loci_df, x="MsCAVIAR", y="CAVIAR-Asian")
ax1.set(xlabel="", ylabel="CAVIAR-Asian Set Sizes")
ax1.set(xlim=(0, 450), ylim=(0, 450))
ax1.plot([0, 1], [0, 1], transform=ax1.transAxes, color='red')

plt.subplot(1, 3, 2)
ax2 = sns.scatterplot(data=hdl_loci_df, x="MsCAVIAR", y="CAVIAR-Euro")
ax2.set(xlabel="MsCAVIAR Set Sizes", ylabel="CAVIAR-Euro Set Sizes")
ax2.set(xlim=(0, 320), ylim=(0, 320))
ax2.plot([0, 1], [0, 1], transform=ax2.transAxes, color='red')

plt.subplot(1, 3, 3)
ax3 = sns.scatterplot(data=hdl_loci_df, x="MsCAVIAR", y="PAINTOR")
ax3.set(xlabel="", ylabel="PAINTOR Set Sizes")
ax3.set(xlim=(0, 300), ylim=(0, 300))
ax3.plot([0, 1], [0, 1], transform=ax3.transAxes, color='red')

plt.xticks(rotation=0)
plt.savefig('hdl_scatter.png', bbox_inches='tight')
```
